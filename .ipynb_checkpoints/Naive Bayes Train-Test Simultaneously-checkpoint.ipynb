# The probability density function for Multivariate Normal is
$$
f(x)=\frac{1}{\sqrt{(2 \pi)^{k} \operatorname{det} \Sigma}} \exp \left(-\frac{1}{2}(x-\mu)^{T} \Sigma^{-1}(x-\mu)\right)
$$

where $\mu$ is the mean, $\Sigma$ the covariance matrix, and **k** is the dimension of the space where x takes values.

```python
def jointPDF(data, mean , cov):
    data = np.array(data)
    mean = np.array(mean)
    cov = np.array(cov)
    a = 1 / (np.sqrt((2*np.pi)**5 * np.linalg.det(cov)))
    b = np.exp( (-1/2) * np.dot(( data - mean ).dot(np.linalg.inv(cov)), ( data - mean ).T))
    return a * b
```
